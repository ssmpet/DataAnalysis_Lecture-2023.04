### 수원 소재 관공서 지도위에 표시하기

```python
import folium
import pandas as pd
import requests, json
from urllib.parse import quote
```

```python
df = pd.read_csv('data/수원소재관공서2.csv')
df['color'] = ['red', 'green', 'blue', 'orange', 'violet']
df
```

```python
map1 = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=13)
for i in df.index:
    folium.Circle(
        location=[df.위도[i], df.경도[i]], 
        popup=folium.Popup(df.주소[i], max_width=200),
        tooltip=df.이름[i],
        color=df.color[i],
        radius=300, fill=True
    ).add_to(map1)

title_html = '<h3 align="center" style="font-size:20px;">수원 소재 주요 관공서</h3>'
map1.get_root().html.add_child(folium.Element(title_html))
map1
```
