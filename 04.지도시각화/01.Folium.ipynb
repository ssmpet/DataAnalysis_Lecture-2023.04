### Folium 지도 시각화

- 설치: pip install folium

```python
import folium
folium.__version__
```

```python
map = folium.Map(location=[37.295, 127.045], zoom_start=14) # Open Street Map
folium.Marker(
    location=[37.2949837, 127.043902], # wGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청'
).add_to(map)
map
```

```python
map = folium.Map(location=[37.295, 127.045], zoom_start=14, tiles='Stamen Toner')
folium.Marker(
    location=[37.2949837, 127.043902], # wGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청'
).add_to(map)
map
```

- 경기도청, 장안구청, 팔달구청, 권선구청, 영통구청

```python
# map = folium.Map(location=[37.2825, 127.0190], zoom_start=13)
map = folium.Map(location=[37.2710, 127.0154], zoom_start=13)

folium.Marker(
    location=[37.2949837, 127.043902], # wGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청',
    icon=folium.Icon(
        color='purple', 
        icon_color='white',
        icon='glyphicon glyphicon-thumbs-up',
        prefix='glyphicon'
    )
).add_to(map)
folium.Marker(
    location=[37.30337225, 127.0096118], 
    popup=folium.Popup('경기도 수원시 장안구 송원로 101', max_width=200),
    icon=folium.Icon(color='green'),
    tooltip='장안구청'
).add_to(map)
folium.Marker(
    location=[37.28256171, 127.0190196],
    popup=folium.Popup('경기도 수원시 팔달구 창룡대로 23', max_width=200),
    tooltip='팔달구청',
    icon=folium.Icon(color='red', icon='cloud')
).add_to(map)
folium.Circle(
    radius=300,
    location=[37.25764489, 126.9719386],
    popup=folium.Popup('경기도 수원시 권선구 호매실로 12', max_width=200),
    tooltip='권선구청',
    color='crimson',
    fill=True
).add_to(map)
folium.CircleMarker(
    location=[37.2596127,127.0465469],
    radius=20,
    popup=folium.Popup('경기도 수원시 영통구 효원로 407', max_width=200),
    tooltip='영통구청',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px;">수원 소재 주요 관공서</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
