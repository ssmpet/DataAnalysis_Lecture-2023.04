```python
titanic = sns.load_dataset("titanic")   # 타이타닉호 데이터
print(titanic.shape)
titanic.head()
```

```python
# 분포
sns.displot(data=titanic, x='age', hue='who', hue_order=['woman', 'man', 'child'])
plt.show()
```

```python
# 히스토그램과 커널밀도 추정함수(kde=True)
sns.displot(data=titanic, x='age', kde=True)
plt.show()
```

```python
sns.histplot(data=titanic, x='age', hue='survived')
plt.show()
```

```python
# 나이별 생존 유무
sns.kdeplot(data=titanic, x='age', hue='alive')
plt.show()
```

```python
# 범주형
sns.boxplot(data=titanic, x='who', y='age')
plt.show()
```

```python
# 탑승자 등급별 운임
sns.barplot(x='class', y='fare', data=titanic)
plt.show()
```

```python
# 탑승자 등급별 나이
sns.boxplot(x='class', y='age', data=titanic)
plt.show()
```

```python
# 탑승자의 등급별 인원수
sns.countplot(x='class', data=titanic)
plt.show()
```
