- Line graph

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randn(100, 3), columns=list('ABC'), 
                  index=pd.date_range('2023-04-01', periods=100)).cumsum()
df.head()
```

```python
df.plot(figsize=(5, 3))       # default가 line graph
plt.title('Pandas의 Plot메소드 사용 예')
plt.xlabel('일자')
plt.ylabel('데이터')
plt.show()
```

- Bar chart

```python
iris = sns.load_dataset('iris')

iris.sepal_length[:30].plot(kind='bar', figsize=(6, 3), rot=45)
plt.title('Iris Setosa 품종의sepal length')
plt.show()
```

```python
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris Setosa 품종의sepal length')
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2
```

```python
df2.plot.bar(rot=0, figsize=(8,5))
# df2.plot(kind='bar')
plt.title('Iris 품종별 Feature별 평균')
plt.show()
```

```python
df2.columns.name = 'Feature'
df2.plot.bar(rot=0, figsize=(6, 3))
plt.ylim(0, 10)                     # 범례가 그림과 겹치지 않게
plt.title('Iris 품종별 Feature별 평균')
plt.show()
```

- Pie chart

```python
titanic = sns.load_dataset('titanic')

s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.1f%%')
plt.title('선실별 승객수')
plt.axis('equal')
plt.show()
```

- histogram, kernel density

```python
iris.plot.hist()
plt.title('Irs 피쳐 Histogram')
plt.show()
plt.title('Iris 피쳐 Kernel Density')
iris.plot.kde()     # scipy 모듈 설치 필요
plt.show()
```

- box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Iris Setosa 품종의 각 피쳐별 Box plot')
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout()
plt.suptitle('Iris 각 피쳐의 품종별 박스 플롯')
plt.show()
```
