1. Tips 데이터
    - 요일별 팁의 비율(bar)
    - 인원수별 팁의 비율(bar)

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = (tips.tip/tips.total_bill*100).round(2)
```

```python
tips.head()
```

```python
tp = tips.pivot_table('tip_pct', 'day')
tp
```

```python
plt.figure(figsize=(4, 3))
plt.bar(tp.index, tp.tip_pct)
plt.title('요일별 팁 비율')
plt.xlabel('요일'); plt.ylabel('팁 비율')
plt.show()
```

```python
stp = tips.pivot_table('tip', 'size')
stp
```

```python
plt.figure(figsize=(5, 3))
plt.bar(stp.index, stp.tip)
plt.title('인원수별 팁 비율')
plt.xlabel('인원수(명)'); plt.ylabel('팁의 비율');
```

3. 서울시 구별 CCTV 댓수 (barh)
    - 기본순서
    - index로 sort (오름)
    - value로 sort (내림)

```python
columns = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
# cctv = pd.read_csv('../99.과제/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr', usecols=columns, thousands=',')
# cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr', usecols=columns, thousands=',')
cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv',  encoding='euc-kr', thousands=',')

# cctv
```

```python
cctv.head()
```

```python
cctv = cctv.iloc[:, 0:13]
# cctv.rename(columns={'2013년 이전\r\n설치된 CCTV':'2013년 이전 설치된 CCTV'})
cctv.columns = columns
cctv.drop(index=0, inplace=True)
cctv.구분 = cctv.구분.str.replace(' ', '')
```

```python
cctv.head()
```

```python
# cctv = cctv[['구분', '총계']]
```

```python
cctv_tolal = cctv[['구분', '총계']].sort_index(ascending=False)

plt.barh(cctv_tolal['구분'], cctv_tolal['총계'])
plt.title('서울시 구별 CCTV대수')
plt.show()
```

```python
cctv_tolal_index = cctv[['구분', '총계']].sort_values('구분', ascending=False)
plt.barh(cctv_tolal_index['구분'], cctv_tolal_index['총계'])
plt.title('자치구별 CCTV 설치대수(자치구명으로 정렬)')
plt.show()
```

```python
cctv_total_val = cctv[['구분', '총계']].sort_values('총계')
plt.barh(cctv_total_val['구분'], cctv_total_val['총계'])
plt.title('자치구별 CCTV 설치대수(대수별로 정렬)')
plt.show()
```
