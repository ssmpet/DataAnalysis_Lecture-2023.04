1. cctv 데이터를 다음과 같이 만들어 cctv.csv 파일에 저장하세요.

   ​       구별    CCTV댓수  최근증가율

        1 종로구 1715         20.69

        2 중구    2447         78.98

        3 용산구 2611          78.64

              ...

​

- 첫번째 행인 계는 삭제한다.

- 최근증가율은 최근 3개년간의 CCTV 댓수의 합을 그 이전 모든 년도의 CCTV 댓수의 합으로 나눈 값으로 백분율 값이다.​

- 구별은 기존의 구분 열을 CCTV댓수는 기존의 총계 열을 이름만 변경한 것이다.

```python

import numpy as np
import pandas as pd
```

```python
columns = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
seoul_cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr', usecols=columns, thousands=',')
seoul_cctv.drop(0, inplace=True)
# cctv = cctv_seoul.iloc[1:]  # 슬라이스로 해도 된다.
```

```python
seoul_cctv.head()
```

```python
seoul_cctv['최근증가율'] = seoul_cctv.apply(lambda x: round(x['2020년':'2022년'].sum() \
                                                  / x['2013년 이전 설치된 CCTV':'2019년'].sum() \
                                                  * 100, 2), axis=1) 
```

```python
cctv = seoul_cctv[['구분', '총계', '최근증가율']]
```

```python
cctv.columns = [['구별', 'CCTV댓수', '최근증가율']]
cctv
```

```python
cctv.to_csv('data/cctv.csv', index=False)
```

2. 서울시 인구 데이터를 다음과 같이 만들어 pop.csv 파일에 저장하세요.


        구별     인구수  내국인  외국인  고령자  외국인비율  고령자비율

      1 종로구   153789   144683    9106    27818   5.92           18.09

      2 중구     131787   122499    9288    24392   7.05           18.51

      3 용산구   237285   222953    14332   39070   6.04           16.47

- 첫번째 행인 합계는 삭제한다.

- 외국인비율은 외국인 수를 총 인구수로 나눈 백분율 값이다.

- 고령자비율은 65세 이상 인구 수를 총 인구수로 나눈 백분율 값이다.

```python
seoul_pop = pd.read_csv('data/서울시인구.txt', sep='\t', skiprows=(0, 3), header=[0,1], thousands=',')
seoul_pop.head()
```

```python
seoul_columns = '구별 인구수 내국인 외국인 고령자'.split()
pop = seoul_pop[[('자치구', '자치구'), ('합계', '계'), ('한국인','계'), ('등록외국인', '계'), ('65세이상고령자', '65세이상고령자')]].copy(True)
# pop = seoul_pop.iloc[:, [1, 3, 6, 9, 13] ].copy(True) # 인덱스로도 각각 가져 올 수 있다.

pop.columns = seoul_columns
pop.head()
```

```python
pop['외국인비율'] = pop.apply(lambda x: round(x['외국인']/x['인구수'] * 100, 2), axis=1) 
pop['고령자비율'] = pop.apply(lambda x: round(x['고령자']/x['인구수'] * 100, 2), axis=1) 
pop.head()
```

```python
pop.to_csv('data/pop.csv', index=False)
```
