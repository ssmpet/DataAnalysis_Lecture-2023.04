- 2. 그룹분석

```python
import seaborn as sns
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

##### 2.1 group_by

- iris data

```python
iris.head(3)
```

```python
iris.tail(3)
```

```python
iris[50:53]
```

```python
# 품종별 항목의 갯수
iris.groupby('species').count()
```

```python
# 품종별 항목의 평균
iris.groupby('species').mean()
```

```python
# 품종별 sepal_length(꽃받침 길이)의 평균 - 결과는 series
iris.groupby('species').mean()['sepal_length']
```

```python
iris[['sepal_length', 'species']].groupby('species').mean()
```

```python
# 품종별 항목의 평균과 표준편차
iris.groupby('species').agg(['mean', 'std'])
```

```python
# 개별 항목의 다양한 통계
iris.groupby('species')['sepal_width'].agg(['mean', 'std', 'min', 'max'])
```

- tips data

```python
print(tips.shape)
tips.head()
```

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
tips.describe()
```

```python
# 성별, 흡연유무별 팁 비율의 평균
tips.groupby(['sex', 'smoker'])[['tip_pct']].mean()
```

```python
# 성별, 흡연유무별 팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['sex', 'smoker'])[['tip_pct']].agg(['count', 'mean', 'min', 'max'])
```

```python
# 요일별, 시간대별 팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['day', 'time'])[['tip_pct']].agg(['count', 'mean', 'min', 'max'])
```

##### 2.2 pivot_table

```python
# 성별, 흡연유무별 팁의 비율
tips.pivot_table('tip_pct', 'sex', 'smoker', 'mean') # values, index, column, aggfunc
```

```python
# 합계 기능
# 성별, 흡연유무별 인원
tips.pivot_table('tip_pct', 'sex', 'smoker', 'count', margins=True, margins_name='계')
```

##### 2.3 groupby vs pivot_table

```python
# 성별 팁 비율의 평균
tips.groupby(['sex'])[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'sex', aggfunc='mean')
```

```python
# mean 은 생략 가능 : 기본값
tips.pivot_table('tip_pct', 'sex')
```

```python
tips.head()
```

```python
# 흡연 유무에 따른 평균 팁 비율과 최소값, 최대값
tips.groupby(['smoker'])['tip_pct'].agg(['mean', 'min', 'max'])
```

```python
tips.pivot_table('tip_pct', 'smoker', aggfunc=['mean', 'min', 'max'])
```

```python
tips.pivot_table('tip_pct', ['day', 'time', 'size'])
```

```python
tips.groupby(['day', 'time', 'size'])[['tip_pct']].agg(['mean', 'std'])
```

```python
tips.groupby('day')['tip_pct'].agg(['mean', 'std'])
```

```python
tips.groupby('time')['tip_pct'].agg(['mean', 'std'])
```

```python
tips.groupby('size')['tip_pct'].agg(['mean', 'std'])
```

- 연습 문제 4.7.4

```python
tt = sns.load_dataset('titanic')
tt.head()
```

```python
# qcut 명령으로 세 개의 나이 그룹을 만든다.
tt['cat'] = pd.qcut(tt.age, 3, labels=['A1', 'A2', 'A3'])
```

```python
tt
```

```python
# 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산
# 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 
# 열에는 선실 인덱스를 사용
tt.groupby(['sex', 'cat', 'pclass'])['survived'].agg(['mean'])
```

```python
tt.pivot_table('survived', ['sex', 'cat'], 'class')
```

```python
tt.pivot_table('survived', 'sex', 'class')
```
