```python
pop = pd.read_csv('data/주민등록인구집계현황.csv', encoding='euc-kr')
pop.head()
```

```python
pop.isna().sum().sum()
```

```python
pop = pop[['행정구역구분명', '행정구역명', '20~29세 (여)', '30~39세 (여)', '60~69세', '70~79세', '80~89세', '90~99세', '100세 이상']]
pop.head()
```

```python
except_city = ['경기도 수원시', '경기도 고양시', '경기도 용인시', '경기도 성남시', '경기도 안산시', '경기도 안양시']
```

```python
# gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
# pop.행정구역명 = pop.행정구역명.str.strip()
# drop_index = []
# for index in pop[pop.행적구역구분명 == '시군'].index:
#     if pop.행정구역명[index].split()[-1] in gu_city:
#         drop_index.append(index)
# drop_index

# pop.drop(drop_index, inplace=True)
# pop.head()
```

```python


pop = pop[((pop.행정구역구분명 == '시군') | (pop.행정구역구분명 == '구')) & ~(pop['행정구역명'].str.strip().isin(except_city))]
pop.head()
```

```python
pop['20-39'] = pop.iloc[:, 2:4].sum(axis=1)
pop['60이상'] = pop.iloc[:, 4:9].sum(axis=1)
pop.head()
# [pop.iloc[:, 1:3].sum(axis=1), pop.iloc[:, 3:].sum(axis=1)]
```

```python
type(pop.iloc[1, 1])
pop['인구소멸비율'] = (pop.iloc[:, 2:4].sum(axis=1) / pop.iloc[:, 4:9].sum(axis=1)).round(4)

# pop[[pop['20~29세 (여)', '30~39세 (여)'].sum(), pop['60~69세 (여)', '70~79세 (여)', '80~89세 (여)', '90~99세 (여)', '100세 이상 (여)'].sum()]]
```

```python
pop = pop[['행정구역명', '인구소멸비율']]
pop.head()
```
