```python
pop = pd.read_csv('data/주민등록인구집계현황.csv', encoding='euc-kr')
pop.head()
```

```python
pop.isna().sum().sum()
```

```python
except_city = ['경기도 수원시', '경기도 고양시', '경기도 용인시', '경기도 성남시', '경기도 안산시', '경기도 안양시']
```

```python
#pop[~(pop['행정구역명'].str.strip().isin(except_city))].head(15)
```

```python

pop = pop[((pop.행정구역구분명 == '시군') | (pop.행정구역구분명 == '구')) & ~(pop['행정구역명'].str.strip().isin(except_city))]
pop.head()
```

```python
pop = pop[['행정구역명', '20~29세 (여)', '30~39세 (여)', '60~69세 (여)', '70~79세 (여)', '80~89세 (여)', '90~99세 (여)', '100세 이상 (여)']]
pop.head()

```

```python
pop['20-39'] = pop.iloc[:, 1:3].sum(axis=1)
pop['60이상'] = pop.iloc[:, 3:8].sum(axis=1)
pop
# [pop.iloc[:, 1:3].sum(axis=1), pop.iloc[:, 3:].sum(axis=1)]
```

```python
type(pop.iloc[1, 1])
pop['인구출산률'] = (pop.iloc[:, 1:3].sum(axis=1) / pop.iloc[:, 3:8].sum(axis=1)).round(4)

# pop[[pop['20~29세 (여)', '30~39세 (여)'].sum(), pop['60~69세 (여)', '70~79세 (여)', '80~89세 (여)', '90~99세 (여)', '100세 이상 (여)'].sum()]]
```

```python
pop
```

```python
pop = pop[['행정구역명', '인구출산률']]
pop.head()
```

```python
pop[pop.인구출산률 <= 0.5].sort_values('인구출산률')
```
