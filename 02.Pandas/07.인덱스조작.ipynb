### 5. 인덱스 조작

```python
import numpy as np
import pandas as pd
```

- 인덱스 설정 및 제거

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,
                    columns=['C1', 'C2', 'C3', 'C4'])
df1
```

```python
df2 = df1.set_index('C1')
df2
```

```python
df1.index, df2.index
```

```python
df3 = df2.set_index('C2')
df3
```

```python
df2.reset_index(inplace=True)
df2
```

```python
# 5명의 학생


df_score1 = pd.DataFrame(np.random.randint(60, 101, size=(5,3)),
                         columns=['국어','영어', '수학'])

df_score1
```

- 다중 인덱스

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])
df3
```

```python
df3.columns.names = ['Cidx1', 'Cidx2']
df3
```

```python
df3['A', 'C1']
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
# 0.41 값을 인덱싱
df4['A', 'D']['M', 'id_3']
```

```python
df4.iloc[2, 1], df4.loc[('M','id_3'), ('A','D')]
```
