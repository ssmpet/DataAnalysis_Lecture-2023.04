- 연습 문제 4.1.3

```python
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
dff = pd.DataFrame(data, index=index, columns=columns)
dff
```

```python
# 모든 학생의 수학 점수를 시리즈로
dff.수학
```

```python
# 모든 학생의 국어와 영어 점수를 데이터 프레임으로
dff[['국어', '영어']]
```

```python
# 모든 학생의 각 과목 평균 점수를 새로운 열로 추가
dff['평균'] = ((dff.국어 + dff.영어 + dff.수학)/3).round(1)
dff
```

```python
# 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산
import warnings
warnings.filterwarnings('ignore')
dff['영어']['방자'] = 80
dff['평균'] = ((dff.국어 + dff.영어 + dff.수학)/3).round(1)
dff
```

```python
# 춘향의 점수를 데이터프레임으로 나타낸다
dff.T[['춘향']]
```
