```python
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('static/img/flower3.jpg')
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
np.array(img).shape
```

```python
h, w, _ = np.array(img).shape
```

```python
if h > w:
    width, height = w, w
else:
    width, height = h, h
```

```python
diff = abs(h-w) // 2
```

```python
if w > h:   # landscap image 가로가 더 큰 경우
    final_img = np.array(img)[:, diff:diff+h, :]
else:
    final_img = np.array(img)[diff:diff+w, :, :]
```

```python
new_img = Image.fromarray(final_img)
plt.imshow(new_img)
plt.axis('off')
plt.show()
```

```python
new_img.save('static/data/profile.png', format='png')
```
